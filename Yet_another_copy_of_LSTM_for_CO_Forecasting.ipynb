```python
#searating CO data from the dataframe for CO2 time series forecasting into a temporary dataframe named temp
temp = df['CO']
temp.plot(figsize=(20,6))
```

```python
#function to append data to X1,y1 for training LSTM
def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(temp, WINDOW_SIZE)
X1.shape, y1.shape
```

```python
#dividing the dataframe into training,validation,testing
X_train1, y_train1 = X1[:5000], y1[:5000]
X_val1, y_val1 = X1[5000:9000], y1[5000:9000]
X_test1, y_test1 = X1[9000:], y1[9000:]
X_train1.shape, y_train1.shape, X_val1.shape, y_val1.shape, X_test1.shape, y_test1.shape
```

```python

#importing keras from tensorflow for using LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

modelJANCombinedCO = Sequential()
modelJANCombinedCO.add(InputLayer((5, 1)))
modelJANCombinedCO.add(LSTM(64))
modelJANCombinedCO.add(Dense(8, 'relu'))
modelJANCombinedCO.add(Dense(1, 'linear'))

modelJANCombinedCO.summary()
```

```python
#adding checkpoint to save the best model while training phase
cp1 = ModelCheckpoint('modelJANCombinedCO/', save_best_only=True)
modelJANCombinedCO.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=['mae','mape',RootMeanSquaredError()])
```

```python
modelJANCombinedCO.fit(X_train1, y_train1, validation_data=(X_val1, y_val1), epochs=500, callbacks=[cp1])
```

```python
# Assuming model is your trained model
modelJANCombinedCO.save('/content/drive/My Drive/my_modell.h5')
```
